## Function for saving match metric to CSV file

```python
def save_display_matches(stitcher, name):
    matches = stitcher.matches
    match_matrix = []
    for i in range(stitcher.num_images):
        out = []
        for match in stitcher.get_idx_slice(i, matches):
            out.append(round(match.confidence, 2))
        match_matrix.append(out)
    
    np.savetxt(name, match_matrix, delimiter=',')
```

## Saving match metric data

```python
save_display_matches(stitcher1, 'match_metric_WLH.csv')
save_display_matches(stitcher2, 'match_metric_office.csv')
save_display_matches(stitcher4, 'match_metric_stjames.csv')
```
